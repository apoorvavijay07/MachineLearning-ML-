8) prog to set font properties for title and labels

23) pie chart with labels

```python
import matplotlib.pyplot as plt
marks=[40,30,20,12]
name=['raju','krishna','ramu','venu']
plt.pie(marks,labels=name,radius=1)
plt.show()
```

24) to display percentage on piechart

```python
import matplotlib.pyplot as plt
marks =[40,30,20,12]
name =['raju','krishna','ramu','venu']
plt.pie(marks,labels=name,radius=1,autopct='%.1f%%')

plt.legend()
plt.show()
```

25)explode

```python
import matplotlib.pyplot as plt
marks =[40,30,20,12]
name =['raju','krishna','ramu','venu']
myexplode = [0,0,0.2,0]
plt.pie(marks,labels=name,radius=1,autopct='%.3f%%',explode=myexplode)

plt.legend()
plt.show()
```
