```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target']=iris.target
df.head()
```

```python
df[df.target==1].head()
```

```python
df[df.target==2].head()
```

```python
df['flower_name']=df.target.apply(lambda x:iris.target_names[x])
df[45:55]
```

```python
df0=df[:50]
df1=df[51:100]
df2=df[101:]
```

```python

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='g',marker='+')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='b',marker='.')
```

```python
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='g',marker='+')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='b',marker='.')
```

SVM

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['target','flower_name'],axis='columns')
y=df.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

REGULARIZATION

```python

model_C=SVC(C=1)
model_C.fit(x_train,y_train)
model_C.score(x_test,y_test)
```

GAMMA

```python
model_g=SVC(gamma=10)
model_g.fit(x_train,y_train)
model_g.score(x_test,y_test)
```

KERNEL

```python
model_linear_kernel=SVC(kernel='linear')
model_linear_kernel.fit(x_train,y_train)
model_linear_kernel.score(x_test,y_test)
```
