13) creating dataframe from list of lists with column and index

```python
import pandas as pd
data = [['a1','b1','c1'],
        ['a2','b2','c2'],
        ['a3','b3','c3']]
columns = ['C1','C2','C3']
index = ['R1','R2','R3']

df=pd.DataFrame(data,index,columns)
print(df)
```
